```python
from transformers import pipeline

# 한국어 GPT-2 모델 로드
generator = pipeline(
    "text-generation",
    model="skt/kogpt2-base-v2",  # SKT 한국어 GPT-2
)

# 텍스트 생성
results = generator(
    "인공지능 기술이 발전하면",   # 시작 문장
    max_length=80,              # 최대 길이
    num_return_sequences=3,     # 3개 생성
    do_sample=True,             # 확률적 샘플링
    temperature=0.8,            # 창의성 조절
)

print("=== 한국어 GPT 텍스트 생성 ===")
for i, r in enumerate(results, 1):
    print(f"\n[생성 {i}]")
    print(r["generated_text"])
```

## 📝 과제

### 과제 1: 한국어 GPT 텍스트 생성

다양한 시작 문장과 temperature로 텍스트를 생성하고 결과를 비교하세요.

```python
prompt = "오늘 저녁으로"

print(f"프롬프트: '{prompt}'\n")

for temp in [0.3, 0.7, 1.2, 1.5]:
    result = generator(
        prompt,
        max_length=60,
        do_sample=True,
        temperature=temp,
        num_return_sequences=1,
    )
    text = result[0]["generated_text"]
    print(f"Temperature={temp}:")
    print(f"  {text}")
    print()

print("0.3: 보수적, 예측 가능")
print("0.7: 적당한 다양성 (추천값)")
print("1.2: 창의적, 엉뚱할 수 있음")
```

### 과제 2: 프롬프트 설계

ChatGPT 또는 Claude에서 아래 3가지를 프롬프트로 해결하세요:
1. 감성 분류 (긍정/부정/중립)
2. 텍스트 요약 (3줄)
3. 키워드 추출 (5개)

각 과제: 프롬프트 설계 → 테스트 → 개선 과정을 기록하세요.

```python
prompts = [
"""
당신은 영화 리뷰 감성 분석 전문가입니다.
리뷰를 읽고 긍정/부정/중립 중 하나로 분류하세요.

예시:
리뷰: 진짜 스케일 하나 진짜 쩌네요 → 긍정
리뷰: 보다가 3번 나갈뻔해서 3점 드립니다 → 부정
리뷰: 좀 난잡한 감이 없지않아 있지만 액션쪽으로 충분히 볼만했던 → 중립

리뷰: 조잡하고 난잡하고 복잡하고 뚜두려깨고 씨끄럽기만했네 →
""",
"""
다음 텍스트를 3줄로 요약하세요.
각 줄은 핵심 내용만 담아주세요.

텍스트: 1,2,3편까진 그래도 먹혔다. 스토리는 무시하고 캐릭터로도 충분히 매력을 보여줬으니까. 4편부터 골수가 빨린 케릭터들은 삐그덕거리기 시작하는데 이번 5편에선 그 정점을 찍었다. 확실한건 캐릭터와 영상미만은 이전만큼 건재하다. 다만 그것만 건재하다.

요약:
""",
"""
다음 텍스트에서 핵심 키워드 5개를 추출하세요.
쉼표로 구분하여 나열하세요.

텍스트: 이제까지 트랜스포머 다 봤는데 이번시리즈는 엄청 지루하더라..계속 지지고 볶으니까 피곤하고 시간기니까 지루하고스토리도 그닥인데다가, CG도 군데군데 이상하더라...게다가 너무 영화화됨 주인공 인간은 잘 안죽고.개연성이 너무 떨어져 몰입안됨

키워드:
"""
]

print("=== 다양한 프롬프트 실험 ===\n")
for prompt in prompts:
    result = generator(
        prompt,
        max_length=50,
        do_sample=True,
        temperature=0.7,
        num_return_sequences=1,
    )
    print(f"입력: '{prompt}'")
    print(f"생성: {result[0]["generated_text"]}")
    print()
```
